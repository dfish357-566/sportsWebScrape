```
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

from time import sleep
from random import randint
```

```
titles = []
shortDesc = []
longDesc = []
simSports = []

sportURL = []
```

```
r = requests.get('https://www.topendsports.com/sport/list/index.htm')
soup = BeautifulSoup(r.text, 'html.parser')
```

```
# Get titles and short descriptions

for li in soup.find_all("li"):
    text = li.text
    # Do not add li after this point, no more sports
    if "Sport by category" in li.text:
      break
    else:
      # Get sports by seperating at "-"
      if li.text.count("—") > 1:
        continue
      elif li.text.count("—") > 1:
        continue
      elif len(li.text) < 1:
        continue
      try:
        #print(li.a.get('href'))
        print(li.text)
        title = text.split("—")[0]
        desc = text.split("—")[1]
        
        sportURL.append(li.a.get('href'))
        titles.append(title)
        shortDesc.append(desc)
        
      except:
        print("Broken")
```

```
# Make sure title corresponds to correct description

print(len(titles))
print(len(shortDesc))

print(titles[810])
print(shortDesc[810])
df = pd.DataFrame({"titles": titles, "description": shortDesc})
#df.to_excel("sportsShortDescriptions.xlsx") 
```

```
print(len(sportURL))
for url in sportURL:
  r = requests.get("https://www.topendsports.com/sport/list/" + url)
  soup = BeautifulSoup(r.text, 'html.parser')
  
  print(soup.find("h1").text)
  #title = soup.find("h1").text
  #titles.append(title)

  longDescription = soup.find_all('p',attrs={'itemprop':None,"class":None})
  #print(longDescription)
  desc = ''
  for item in longDescription:
    if "Share:" not in item.text:
      desc += item.text
    else:
      break
  
  longDesc.append(desc)

  simSportlist = []

  for li.a in soup.find_all("li"):
    if "List of" not in li.a.text:
      simSportlist.append(li.a.text)
      #print(li.a.text)
    else:
      break

  simSports.append(simSportlist)
```
